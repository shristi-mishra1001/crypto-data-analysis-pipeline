```python
df.groupby('crypto_name')['close'].mean().sort_values(ascending=False)
```

```python
btc_df = df[df['crypto_name'] == 'Bitcoin']
btc_df.head()
```

```python
btc_df['close'].max(), btc_df['close'].min()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(btc_df['date'], btc_df['close'])
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Bitcoin Close Price Over Time')
plt.show()
```
